```python
 
df = pesquisa['Sent'].value_counts(normalize=True).reset_index()
df.columns = ['Sentimento', 'Porcentagem']

# Criar o gr√°fico de barras
ax = sns.barplot(x='Sentimento', y='Porcentagem', data=df)

# Adicionar as porcentagens nas barras
ax.bar_label(ax.containers[0], fmt='%.2f%%')

# Mostrar o gr√°fico
sns.boxplot()

```

```python
# Cria um histograma para cada estilo musical
plt.figure(figsize=(14, 12))

for i, column in enumerate(df_nota.columns, 1):
    plt.subplot(3, 3, i)  # Ajusta o layout para v√°rios subgr√°ficos
    sns.histplot(df_nota[column], bins=5, kde=True, color='skyblue')
    plt.title(f'Estilo: {column}')
    plt.xlabel('Nota')
    plt.ylabel('Frequ√™ncia')

plt.tight_layout()
plt.show()
```
