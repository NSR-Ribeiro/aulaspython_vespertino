```python
import pandas as pd # Importanto o Pandas
import matplotlib.pyplot as plt # Importando o Matplotlib
import seaborn as sns # Importanto o Seaborn 
```

```python
df = pd.read_csv('dataset/pesquisa_musical_alunos.csv') # Lendo o Arquivo
```

```python
df.head() # Mostrando os Primeiros
```

```python
df.columns = ['data_hora','rock','rap','mpb','reggae','funk','sertanejo','gospel','eletronica','professor','faixa_etaria','samba','forro','trap','palavra_chave']
# Renomeando as Colunas
```

```python
df #mostrando a tabela
```

```python
df_estilos = df[['rock', 'rap', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica','samba', 'trap', 'forro']]
# Criando uma tabela apenas para os estilos 
```

```python
df_estilos 
```

```python
df.describe() #Mostrando todos os valores de calculos Estatística
```

```python
#Somando os Valores da tabela
df_estilos_soma = df_estilos[['rock', 'rap', 'mpb', 'reggae', 'funk', 'sertanejo', 'gospel', 'eletronica','samba', 'trap', 'forro']].sum() 
df_estilos_soma
```

```python
df_estilos_soma.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Quantidade de Ocorrências por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)  # Rotaciona os nomes das categorias para melhor leitura
plt.show()
```

```python
# Crie uma série com os valores dos estilos musicais
estilos_musicais = pd.Series({
    'rock': 3.251701,
    'rap': 2.986395,
    'mpb': 3.061224,
    'reggae': 2.455782,
    'funk': 2.462585,
    'sertanejo': 2.850340,
    'gospel': 2.619048,
    'eletronica': 3.278912
})

# Calcula a média
media = estilos_musicais.mean()

print(f"A média dos estilos musicais é: {media:.2f}")
```

```python
df_estilos_preenchidos = df_estilos.fillna(2.87)  # Preenche com a media
```

```python
df_estilos_preenchidos # Mostrando a tabela com valores preenchidos
```

```python
df_estilos_preenchidos_soma = df_estilos_preenchidos.sum()
```

```python
df_estilos_preenchidos_soma.plot(kind='bar',figsize=(10, 6), color='skyblue')
plt.title('Quantidade de Ocorrências por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)  # Rotaciona os nomes das categorias para melhor leitura
plt.show()
```
